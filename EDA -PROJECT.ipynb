# Inventory Turnover Rate analysis (Top-20)
What is Inventory Turnover Rate?
-It tells you how many times a product or inventory item is sold and replaced (or "turned over") during a specific time period.

1. High	Product sells quickly; good demand; low risk of overstocking
2. Low	Product moves slowly; might be overstocked or not selling well
                                           

```python

top_turnover = data_1.sort_values(by='Inventory_Turnover_Rate', ascending=False).head(20)
top_turnover
```

```python
low_turnover = data_1.sort_values(by='Inventory_Turnover_Rate').head(20)
low_turnover
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16,6))

sns.barplot(data=top_turnover, y='Product_Name', x='Inventory_Turnover_Rate', ax=axs[0], palette='Greens_r')
axs[0].set_title('Top 20 Fast-Moving Products')

sns.barplot(data=low_turnover, y='Product_Name', x='Inventory_Turnover_Rate', ax=axs[1], palette='Reds')
axs[1].set_title('Top 20 Slow-Moving Products')


plt.show()
```

# CATEGORY WISE SEASONAL DEMAND PATTERN

```python


data_1['Date_Received'] = pd.to_datetime(data_1['Date_Received'])

data_1['Received_Month'] = data_1['Date_Received'].dt.month_name()
```

```python

category_month_sales = data_1.groupby(['Catagory', 'Received_Month'])['Sales_Volume'].sum().reset_index()


month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

category_month_sales['Received_Month'] = pd.Categorical(category_month_sales['Received_Month'],categories=month_order,ordered=True)


plt.figure(figsize=(14, 7))
sns.lineplot(data=category_month_sales, x='Received_Month', y='Sales_Volume', hue='Catagory',marker='o')

plt.title(' Category-wise Seasonal Demand Pattern')
plt.xlabel('Stock Received Month')
plt.ylabel('Total Sales Volume')
plt.legend(title='Category')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Interpretations from seasonal category demand pattern
1. Bakery
   . The highest sales volume or consumption of bakery is recorded in the month of March followed by july.
   . The least sales volume or consumption is recorded in the month of November.

2. Beverages
   . The highest sales volume or consumption of beverages is recoreded in the month of march followed by december.
   . The lowest sales volume or consumption is recorded in the month february.

3. Dairy 
   . The highest sales is recorded in the month of April.
   . The lowest sales is recorded in the month of March.

4. Fruit and Vegetables
   . The highest sales is recorde in the month of july followed by october.
   . The lowest sales is recorded in the month of september followed by february.

5. Grain and Pulses 
   . The highest sales is recorded in the month of october followed by august.
   . The lowest sales is recorded in the month of june

6. Oils and Fats
   . The highest sales is recorded in the month of june.
   . The lowest sales is recorded in the month of february.

7. Seafood
   . The highest sales is recored in the month of february,march and september with slight differences.
   . The lowest sales is recorded in the month of november.

# MONTH-WISE EXPIRING STOCK ANALYSIS


```python
#step 1: convert expiry date into date-time format 
#step 2: find expiry month
#step 3: groupby expiry month and stock quantity

data_1['Expiry_Date'] = pd.to_datetime(data_1['Expiration_Date'])

data_1['Expiry_Month'] = data_1['Expiry_Date'].dt.to_period('M')

Monthly_expiring_stock = data_1.groupby('Expiry_Month')['Stock_Quantity'].sum()

Monthly_expiring_stock.plot(kind='bar',color='pink')
```
